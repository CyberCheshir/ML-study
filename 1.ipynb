Линейная регрессия

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
def show_scatter(x,y,title='Scatter'):    
    plt.title('Scatter')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.scatter(x,y)
    plt.grid()
    plt.show()
```

```python
x = np.array([5, 15, 25, 35, 45, 55],dtype=object).reshape((-1, 1))
y = np.array([28, 11, 2, 9, 29.5, 30],dtype=object)
show_scatter(x,y)
```

```python
model = LinearRegression().fit(x, y)
r_sq = model.score(x, y) # coef determination predict
b0, b1 = model.intercept_,model.coef_
f = x*b1+b0
```

```python
def mse_lines_regress(x,y,f):
    for i in range(len(x)):
        plt.plot([x[i],x[i]],[f[i],y[i]],color="red")

plt.title(f'Score: {model.score}')
mse_lines_regress(x,y,f)
plt.plot(x,f)
show_scatter(x,y)
```

Полиноминальная регрессия 

```python
import random
import seaborn as sns
```

```python
def random_points(n=100):
    points_x, points_y = [], []
    for i in range(n):
        points_x.append(random.random())
        points_y.append(random.random())
    return {'x': points_x, 'y': points_y}

# ИДЕЯ: как рисовать не затрагивая оперативу?
# Отрисовать точки в зависимости от функции

points = random_points(n=10**3)
sns.jointplot(data=points,x='x',y='y')
```
